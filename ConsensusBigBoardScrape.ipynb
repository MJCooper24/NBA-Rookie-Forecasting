```python
import pandas as pd
from fuzzywuzzy import fuzz
import re

# Step 1: Create consensus board data from the scraped information
# This data is from the Rookie Scale website as of the last update
consensus_data = [
    {'Rank': '1', 'Prospect': 'Cooper Flagg'},
    {'Rank': '2', 'Prospect': 'Dylan Harper'},
    {'Rank': '5', 'Prospect': 'Ace Bailey'},
    {'Rank': '3', 'Prospect': 'VJ Edgecombe'},
    {'Rank': '12', 'Prospect': 'Kasparas Jakučionis'},
    {'Rank': '7', 'Prospect': 'Tre Johnson'},
    {'Rank': '4', 'Prospect': 'Kon Knueppel'},
    {'Rank': '8', 'Prospect': 'Khaman Maluach'},
    {'Rank': '13', 'Prospect': 'Derik Queen'},
    {'Rank': '6', 'Prospect': 'Jeremiah Fears'},
    {'Rank': '23', 'Prospect': 'Liam McNeeley'},
    {'Rank': '18', 'Prospect': 'Asa Newell'},
    {'Rank': '11', 'Prospect': 'Collin Murray-Boyles'},
    {'Rank': '14', 'Prospect': 'Egor Demin'},
    {'Rank': '22', 'Prospect': 'Nolan Traoré'},
    {'Rank': '9', 'Prospect': 'Noa Essengue'},
    {'Rank': '25', 'Prospect': 'Jase Richardson'},
    {'Rank': '32', 'Prospect': 'Ben Saraf'},
    {'Rank': '27', 'Prospect': 'Rasheer Fleming'},
    {'Rank': '17', 'Prospect': 'Thomas Sorber'},
    {'Rank': '19', 'Prospect': 'Danny Wolf'},
    {'Rank': '26', 'Prospect': 'Hugo González'},
    {'Rank': '29', 'Prospect': 'Noah Penda'},
    {'Rank': '24', 'Prospect': 'Will Riley'},
    {'Rank': '10', 'Prospect': 'Carter Bryant'},
    {'Rank': '55', 'Prospect': 'Ryan Nembhard'},
    {'Rank': '44', 'Prospect': 'Koby Brea'},
    {'Rank': '33', 'Prospect': 'Adou Thiero'},
    {'Rank': '39', 'Prospect': 'Kam Jones'},
    {'Rank': '48', 'Prospect': 'Alijah Martin'},
    {'Rank': '46', 'Prospect': 'John Tonje'},
    {'Rank': '47', 'Prospect': 'Javon Small'},
    {'Rank': '31', 'Prospect': 'Ryan Kalkbrenner'},
    {'Rank': '40', 'Prospect': 'Johni Broome'},
    {'Rank': '20', 'Prospect': 'Nique Clifford'},
    {'Rank': '62', 'Prospect': 'Mark Sears'},
    {'Rank': '58', 'Prospect': 'Dink Pate'},
    {'Rank': '35', 'Prospect': 'Tyrese Proctor'},
    {'Rank': '49', 'Prospect': 'Micah Peavy'},
    {'Rank': '30', 'Prospect': 'Drake Powell'},
    {'Rank': '37', 'Prospect': 'Chaz Lanier'},
    {'Rank': '45', 'Prospect': 'Rocco Zikarsky'},
    {'Rank': '43', 'Prospect': 'Alex Toohey'},
    {'Rank': '16', 'Prospect': 'Joan Beringer'},
    {'Rank': '50', 'Prospect': 'Hunter Sallis'},
    {'Rank': '51', 'Prospect': 'Kobe Sanders'},
    {'Rank': '41', 'Prospect': 'Sion James'},
    {'Rank': '28', 'Prospect': 'Maxime Raynaud'},
    {'Rank': '60', 'Prospect': 'Brice Williams'},
    {'Rank': '54', 'Prospect': 'Lachlan Olbrich'},
    {'Rank': '15', 'Prospect': 'Cedric Coward'},
    {'Rank': '36', 'Prospect': 'Yanic Konan Niederhäuser'},
    {'Rank': '21', 'Prospect': 'Walter Clayton Jr.'},
    {'Rank': '42', 'Prospect': 'Jamir Watkins'},
    {'Rank': '38', 'Prospect': 'Bogoljub Marković'},
    {'Rank': '56', 'Prospect': 'Vlad Goldin'},
    {'Rank': '52', 'Prospect': 'Eric Dixon'},
    {'Rank': '34', 'Prospect': 'Hansen Yang'},
    {'Rank': '53', 'Prospect': 'Amari Williams'},
    {'Rank': '57', 'Prospect': 'RJ Luis Jr.'},
    {'Rank': '63', 'Prospect': 'Max Shulga'},
    {'Rank': '61', 'Prospect': 'Izan Almansa'},
    {'Rank': '59', 'Prospect': 'Tamar Bates'},
    {'Rank': '65', 'Prospect': 'Grant Nelson'},
    {'Rank': '70', 'Prospect': 'Payton Sandfort'},
    {'Rank': '67', 'Prospect': 'Will Richard'},
    {'Rank': '69', 'Prospect': 'Jax Robinson'},
    {'Rank': '71', 'Prospect': 'Caleb Love'},
    {'Rank': '64', 'Prospect': 'Chucky Hepburn'},
    {'Rank': '72', 'Prospect': 'Kobe Johnson'},
    {'Rank': '66', 'Prospect': 'Brooks Barnhizer'},
    {'Rank': '73', 'Prospect': 'Igor Miličić Jr.'},
    {'Rank': '68', 'Prospect': 'Viktor Lakhin'},
    {'Rank': '75', 'Prospect': 'Arthur Kaluma'},
    {'Rank': '75', 'Prospect': 'Kadary Richmond'},
    {'Rank': '82', 'Prospect': 'Jalon Moore'},
    {'Rank': '80', 'Prospect': 'Matthew Cleveland'},
    {'Rank': '77', 'Prospect': 'John Poulakidas'},
    {'Rank': '78', 'Prospect': 'Mohamed Diawara'},
    {'Rank': '84', 'Prospect': 'Coleman Hawkins'},
    {'Rank': '91', 'Prospect': 'Wooga Poplar'},
    {'Rank': '85', 'Prospect': 'Jacksen Moni'},
    {'Rank': '89', 'Prospect': 'Dawson Garcia'},
    {'Rank': '76', 'Prospect': 'Curtis Jones'},
    {'Rank': '83', 'Prospect': 'Dylan Cardwell'},
    {'Rank': '81', 'Prospect': 'Sean Pedulla'},
    {'Rank': '88', 'Prospect': 'Eli John N\'Diaye'},
    {'Rank': '87', 'Prospect': 'Lamont Butler'},
    {'Rank': '86', 'Prospect': 'Caleb Grill'},
    {'Rank': '94', 'Prospect': 'Adama-Alpha Bal'},
    {'Rank': '95', 'Prospect': 'Matt Murrell'},
    {'Rank': '93', 'Prospect': 'Miles Kelly'},
    {'Rank': '90', 'Prospect': 'Jahmai Mashack'},
    {'Rank': '96', 'Prospect': 'Norchad Omier'},
    {'Rank': '79', 'Prospect': 'Saliou Niang'},
    {'Rank': '92', 'Prospect': 'Cliff Omoruyi'},
    {'Rank': '97', 'Prospect': 'Andrew Carr'},
    {'Rank': '98', 'Prospect': 'Isaac Nogués'},
    {'Rank': '99', 'Prospect': 'RJ Davis'},
    {'Rank': '100', 'Prospect': 'Chance McMillian'},
    {'Rank': 'NR', 'Prospect': 'Taelon Peter'}
]

consensus_df = pd.DataFrame(consensus_data)

# Step 2: Load your existing CSV
draft_df = pd.read_csv('2025_draft_final.csv')

print(f"Loaded {len(draft_df)} players from your CSV")
print(f"Loaded {len(consensus_df)} players from consensus board")

# Step 3: Function to clean player names for better matching
def clean_name(name):
    """Remove Jr., Sr., II, III, etc. and normalize spacing"""
    if pd.isna(name):
        return ""
    name = str(name)
    # Remove suffixes
    name = re.sub(r'\s+(Jr\.?|Sr\.?|II|III|IV)$', '', name, flags=re.IGNORECASE)
    # Normalize spacing
    name = ' '.join(name.split())
    return name.strip()

# Clean names in both dataframes
consensus_df['Prospect_Clean'] = consensus_df['Prospect'].apply(clean_name)

# Get the player name column (assumes it's the first column or has 'name' in it)
player_col = draft_df.columns[0]
for col in draft_df.columns:
    if 'name' in col.lower() or 'player' in col.lower():
        player_col = col
        break

print(f"Using '{player_col}' as the player name column")

draft_df['Player_Clean'] = draft_df[player_col].apply(clean_name)

# Step 4: Match players using fuzzy matching
def find_best_match(player_name, consensus_df, threshold=85):
    """Find best matching player name using fuzzy string matching"""
    best_match = None
    best_score = 0
    best_idx = None
    
    for idx, row in consensus_df.iterrows():
        consensus_name = row['Prospect_Clean']
        score = fuzz.ratio(player_name.lower(), consensus_name.lower())
        if score > best_score and score >= threshold:
            best_score = score
            best_match = row['Prospect']
            best_idx = idx
    
    return best_match, best_score, consensus_df.loc[best_idx, 'Rank'] if best_idx is not None else None

# Create a mapping of consensus ranks
consensus_rank_map = {}
for idx, row in consensus_df.iterrows():
    consensus_rank_map[row['Prospect_Clean']] = row['Rank']

# Add consensus rank column
draft_df['Consensus_Big_Board_Rank'] = None
matched_players = []

for idx, row in draft_df.iterrows():
    player_name = row['Player_Clean']
    
    # Try exact match first
    if player_name in consensus_rank_map:
        rank = consensus_rank_map[player_name]
        draft_df.at[idx, 'Consensus_Big_Board_Rank'] = rank
        matched_players.append((player_name, player_name, 100, rank))
    else:
        # Try fuzzy matching
        matched_name, score, rank = find_best_match(player_name, consensus_df)
        if matched_name is not None:
            draft_df.at[idx, 'Consensus_Big_Board_Rank'] = rank
            matched_players.append((player_name, matched_name, score, rank))
            if score < 100:
                print(f"Fuzzy matched: '{player_name}' -> '{matched_name}' (score: {score}, rank: {rank})")

# Drop the temporary clean name column
draft_df = draft_df.drop('Player_Clean', axis=1)

# Step 5: Save the updated CSV
output_file = '2025_draft_final_with_consensus.csv'
draft_df.to_csv(output_file, index=False)

print(f"\n✓ Successfully added consensus ranks")
print(f"✓ Saved to: {output_file}")

# Show summary statistics
matched = draft_df['Consensus_Big_Board_Rank'].notna().sum()
total = len(draft_df)
print(f"\nMatched {matched} out of {total} players ({matched/total*100:.1f}%)")

# Show some examples
print("\nFirst few rows with consensus ranks:")
print(draft_df[[player_col, 'Consensus_Big_Board_Rank']].head(10))
```
